```python
from pathlib import Path
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
data_root = Path("../data/GDBDv2/").resolve()
train_df_path = data_root.joinpath("train_metadata.csv")
val_df_path = data_root.joinpath("val_metadata.csv")
```

```python
train_df = pd.read_csv(train_df_path, index_col=0)
```

```python
train_df_el = train_df.iloc[np.random.randint(0, 10416)]
```

```python
train_df_el
```

```python
train_df_el.gt_img_path
```

```python
gt_image = np.asarray(Image.open(data_root.joinpath(train_df_el.gt_img_path)).convert('L'))
noisy_image = np.asarray(Image.open(data_root.joinpath(train_df_el.noisy_img_path)).convert('L'))
```

```python
_, ax = plt.subplots(1,2, figsize=(20,10))
ax[0].imshow(gt_image, cmap='gray')
ax[0].title.set_text('GT Image')

ax[1].imshow(noisy_image, cmap='gray')
ax[1].title.set_text('Noisy Image')
plt.show()
```
